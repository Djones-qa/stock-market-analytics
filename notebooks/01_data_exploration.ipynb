# 01 - Data Exploration
Load raw stock data, validate schema, explore structure.

```python
import sys
sys.path.append('..')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.data_loader import load_raw_data, validate_schema, load_config
from src.utils import print_summary
config = load_config('../config/config.yaml')
```

```python
df = load_raw_data('prices.csv', config)
print_summary(df, 'Raw Stock Data')
```

```python
schema = validate_schema(df, config)
print(f"Valid: {schema['valid']}")
print(f"Missing: {schema['missing']}")
```

```python
fig, ax = plt.subplots(figsize=(14, 6))
sns.heatmap(df.isna(), cbar=True, yticklabels=False, cmap='YlOrRd', ax=ax)
ax.set_title('Missing Values', fontweight='bold')
plt.tight_layout()
plt.show()
```
